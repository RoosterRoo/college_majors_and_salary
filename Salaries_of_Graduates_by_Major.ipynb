What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

Answer: Chemical Engineers have the highest mid-career salary and they earn 107000 on an average

Which college major has the lowest starting salary and how much do graduates earn after university?

```python
lowest_starting_salary_major_idx = na_clean_df['Starting Median Salary'].idxmin()
lowest_starting_salary_major_idx
```

```python
lowest_starting_salary_major = na_clean_df.loc[lowest_starting_salary_major_idx]['Undergraduate Major']
lowest_starting_salary_major
```

```python
lowest_starting_salary_major_earnings = na_clean_df.loc[lowest_starting_salary_major_idx]['Starting Median Salary']
lowest_starting_salary_major_earnings
```

Answer: Spanish Majors have the lowest starting salary and they earn 34000.0 on an average

Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? 

```python
lowest_mid_career_salary_idx = na_clean_df['Mid-Career Median Salary'].idxmin()
lowest_mid_career_salary_major = na_clean_df.loc[lowest_mid_career_salary_idx]['Undergraduate Major']
lowest_mid_career_salary_major
```

```python
lowest_mid_career_salary = na_clean_df.loc[lowest_mid_career_salary_idx]['Mid-Career Median Salary']
lowest_mid_career_salary
```

Answer: Education Majors have the lowest mid-career salary earning 52000

**Majors with the Greatest Spread in Salaries**

```python
spread = na_clean_df['Mid-Career 90th Percentile Salary'].subtract(na_clean_df['Mid-Career 10th Percentile Salary'])
```

```python
na_clean_df.insert(1, "Spread", spread)
na_clean_df.head()
```

```python
low_risk_majors = na_clean_df.sort_values('Spread')
low_risk_majors[['Undergraduate Major', 'Spread']].head()
```

```python
high_risk_majors = na_clean_df.sort_values('Spread', ascending=False)
high_risk_majors[['Undergraduate Major', 'Spread']].head()
```

**Majors with the Highest Potential**

```python
highest_90th_percentile = na_clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_90th_percentile[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```
